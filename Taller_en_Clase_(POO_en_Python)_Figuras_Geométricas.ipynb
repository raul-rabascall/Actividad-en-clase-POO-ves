```python
"""
Taller Figuras Geométricas

Aplicación de:
- Encapsulamiento con @property
- Atributos privados (__)
- Herencia múltiple
- Sobrescritura de métodos
- Cálculo de área y perímetro
- Validaciones


"""

import math
```

```python
class FiguraGeometrica:
    """Clase base para figuras geométricas."""

    def __init__(self, alto, ancho):
        self.alto = alto
        self.ancho = ancho

    @property
    def alto(self):
        return self.__alto

    @alto.setter
    def alto(self, valor):
        if valor > 0:
            self.__alto = valor
        else:
            raise ValueError("El alto debe ser positivo")

    @property
    def ancho(self):
        return self.__ancho

    @ancho.setter
    def ancho(self, valor):
        if valor > 0:
            self.__ancho = valor
        else:
            raise ValueError("El ancho debe ser positivo")

    def __str__(self):
        return f"Alto: {self.__alto}, Ancho: {self.__ancho}"
```

```python
class Color:
    """Clase que representa el color."""

    def __init__(self, color):
        self.color = color

    @property
    def color(self):
        return self.__color

    @color.setter
    def color(self, valor):
        if valor:
            self.__color = valor
        else:
            raise ValueError("El color no puede estar vacío")

    def __str__(self):
        return f"Color: {self.__color}"
```

```python
class Cuadrado(FiguraGeometrica, Color):
    """Clase que representa un cuadrado."""

    def __init__(self, lado, color):
        if lado <= 0:
            raise ValueError("El lado debe ser positivo")

        FiguraGeometrica.__init__(self, lado, lado)
        Color.__init__(self, color)

    def area(self):
        return self.alto ** 2

    def perimetro(self):
        return 4 * self.alto

    def __str__(self):
        return (
            f"Cuadrado -> {FiguraGeometrica.__str__(self)}, "
            f"{Color.__str__(self)}, "
            f"Área: {self.area()}, Perímetro: {self.perimetro()}"
        )
```

```python
class Rectangulo(FiguraGeometrica, Color):
    """Clase que representa un rectángulo."""

    def __init__(self, alto, ancho, color):
        FiguraGeometrica.__init__(self, alto, ancho)
        Color.__init__(self, color)

    def area(self):
        return self.alto * self.ancho

    def perimetro(self):
        return 2 * (self.alto + self.ancho)

    def __str__(self):
        return (
            f"Rectángulo -> {FiguraGeometrica.__str__(self)}, "
            f"{Color.__str__(self)}, "
            f"Área: {self.area()}, Perímetro: {self.perimetro()}"
        )
```

```python
class Circunferencia(FiguraGeometrica, Color):
    """Clase que representa una circunferencia."""

    def __init__(self, radio, color):
        if radio <= 0:
            raise ValueError("El radio debe ser positivo")

        FiguraGeometrica.__init__(self, radio, radio)
        Color.__init__(self, color)

    def area(self):
        return math.pi * self.alto ** 2

    def perimetro(self):
        return 2 * math.pi * self.alto

    def __str__(self):
        return (
            f"Circunferencia -> Radio: {self.alto}, "
            f"{Color.__str__(self)}, "
            f"Área: {self.area():.2f}, "
            f"Perímetro: {self.perimetro():.2f}"
        )
```

```python
if __name__ == "__main__":
    print("=== PRUEBAS DEL SISTEMA ===\n")

    try:
        cuadrado = Cuadrado(5, "Rojo")
        print(cuadrado)
        print()

        rectangulo = Rectangulo(4, 6, "Azul")
        print(rectangulo)
        print()

        circunferencia = Circunferencia(3, "Verde")
        print(circunferencia)
        print()

    except ValueError as error:
        print("Error:", error)
```
